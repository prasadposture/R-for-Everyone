# Pie Chart
A pie-chart is a representation of values as slices of a circle with different colors.

### Creating a pie-chart
Pie-chart is created using the `pie()` function which takes positive numbers as a vector
input.  
The syntax is `pie(x, labels, radius, main, col, clockwise)` where,
1. **x** is a vector containing the numeric values used in the pie chart.
2. **labels** is used to give description to the slices.
3. **radius** indicates the radius of the circle of the pie chart.(value between âˆ’1 and +1).
4. **main** indicates the title of the chart.
5. **col** indicates the color palette.
6. **clockwise** is a logical value indicating if the slices are drawn clockwise or anti clockwise.

```r
#Example
x=c(25,25,25,25)
#labels the partition
labels=c("Chennai", "Delhi", "Kolkata", "Mumbai")
#Plot the pie chart
pie(x, labels)
```

### Giving title and color to the pie-chart
We use parameter main to add a title to the chart and col parameter to add color to
the pie chart where number of elements in a vector and color vector is same.

```r
#Example
x=c(25,25,25,25)
#labels the partition
labels=c("Chennai", "Delhi", "Kolkata", "Mumbai")
c=c("red", "blue", "green", "orange")
#plot the pie chart
pie(x, labels, main = "City Pie Chart", col = c)
```

**Remark :** one can also use rainbow color pallet while drawing the chart. The length of the pallet
should be same as the number of values we have for the chart. Hence we use length(x).

```r
#Example
pie(x, labels, main ="City Pie Chart",col = rainbow(length(x)))
```

### Adding percentage and legend to the pie-chart
We can add slice percentage using `p=round(100*x/sum(x), 1)` and legend using `legend()`

```r
#percentage
p=round(100*x/sum(x), 1)
#plotting pie chart
pie(x, labels = p, main = "City pie chart", col = rainbow(length(x)))
#adding legend
legend("topright", labels, cex = 0.8,fill = rainbow(length(x)))
```

### 3D Pie Charts
The package plotrix has a function called `pie3D()` that is used for 3D pie chart. For that we need to install `plotrix` in our workspace.

```r
#installing plotrix
install.packages('plotrix')
```

```r
#Example
x=c(25,25,25,25)
#labels the partition
labels=c("Chennai", "Delhi", "Kolkata", "Mumbai")
#colors
c=c("red", "blue", "green", "white")
#importing the library 'plotrix'
library(plotrix)
#ploting the 3D pie chart
pie3D(x, labels = labels, explode = 0.1, main = "3D Pie Chart of Cities ")
```

**Note:** The value of explode denotes the spacing between each block of 3D pie chart. If we put
explode = 0 that is we use, we get following results.

```r
#Example
pie3D(x, labels = labels, explode = 0, main = "3D Pie Chart of Cities ")
```
