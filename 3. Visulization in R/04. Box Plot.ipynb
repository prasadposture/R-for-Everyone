# Box Plot
Boxplots are a measure of how well distributed is the data in a data set. It divides the
data set into three quartiles.

### Graph of box
This graph represents the minimum, maximum, median, first quartile and third quartile
in the data set.

### Creating Box Plot
Boxplots are created in R by using the `boxplot()` function.
The basic syntax to create a
`boxplot in R is boxplot(x, data, notch, varwidth, names, main)`, where,
1. **x** is a vector or a formula.
2. **data** is the data frame.
3. **notch** is a logical value. Set as TRUE to draw a notch.
4. **varwidth** is a logical value. Set as true to draw width of the box proportionate to the sample size.
5. **names** are the group labels which will be printed under each boxplot.
6. **main** is used to give a title to the graph.

**Note:** Type `data()` to know pre loaded data set in you R and then use type R data set to
check available data options in each package.

```r
#Example: We will use 'cars' dataset which is already present in R
i=cars[c("speed","dist")]
print(head(i,5))
```

```r
#Creating box plot
boxplot(speed ~ dist, data = cars, xlab = "speed", ylab = "dist", main = "Speed-Distance")
```

### Box Plot with Notch

```r
#using mtcars data
boxplot(mpg ~ cyl, data = mtcars, 
        xlab = "Number of Cylinders", 
        ylab = "Miles Per Gallon", 
        main = "Mileage Data", 
        notch = TRUE, #plotting with notch
        varwidth = TRUE,
        col = c("orange","white","green"),
        names = c("High","Medium","Low")
       )
```

```r
#Plotting without notch
#using mtcars data
boxplot(mpg ~ cyl, data = mtcars, 
        xlab = "Number of Cylinders", 
        ylab = "Miles Per Gallon", 
        main = "Mileage Data", 
        notch = FALSE,
        varwidth = TRUE,
        col = c("orange","white","green"),
        names = c("High","Medium","Low")
       )
```
