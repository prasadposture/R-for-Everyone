# Histogram
A histogram represents the frequencies of values of a variable bucketed into ranges.
Histogram is similar to bar chat but the difference is it groups the values into continuous ranges.
Each bar in histogram represents the height of the number of values present in that range.

### Creating a histogram
R creates histogram using `hist()` function. The basic syntax for creating a histogram
using R is `hist(V, main, xlab, xlim, ylim, breaks, col, border)`  
Following is the description of the parameters used â€“
1. **V** is a vector containing numeric values used in histogram.
2. **main** indicates title of the chart.
3. **col** is used to set color of the bars.
4. **border** is used to set border color of each bar.
5. **xlab** is used to give description of x-axis.
6. **xlim** is used to specify the range of values on the x-axis.
7. **ylim** is used to specify the range of values on the y-axis.
8. **breaks** is used to mention the width of each bar.


```r
#Example
 # Create data for the graph.
v=c(9,13,21,8,36,22,12,40,31,33,19)
# Give the chart file a name
hist(v, xlab="Weight", col="green", border="red", xlim=c(0,40), ylim=c(0,3), breaks =5)
```

**Note:** To specify the range of values allowed in X axis and Y axis, we can use the xlim and ylim
parameters. The width of each of the bar can be decided by using breaks.

```r
#Example
v=c(2,3,3,8,13,11,14,21,18,34,22,26,12,31,33,19)
c=c("purple", "violet", "blue", "green", "yellow", "orange", "red")
hist(v, xlab = "Weight", col = c, border = "red", xlim = c(0,35), ylim = c(0,5),breaks = 5)
```
