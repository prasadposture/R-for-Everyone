# Bar Plot
A bar chart is used to represent data in rectangular bars with length of the bar
proportional to the value of the variable. 

### Creating a bar chart
R uses the function `barplot()` to create bar charts. The syntax used is
`barplot(H, xlab, ylab, main, names.arg, col)` where,
1. **H** is a vector or matrix containing numeric values used in bar chart.
2. **xlab** is the label for x axis
3. **ylab** is the label for y axis.
4. **main** is the title of the bar chart.
5. **names.arg** is a vector of names appearing under each bar.
6. **col** is used to give colors to the bars in the graph.

```r
#Example : Monthly Revenue of a company
#Creating the data for the chart
H=c(7,1,2,13,14,1,13)
M=c("Mar", "Apr", "May", "Jun", "July", "Aug", "Sept")
barplot(H, names.arg = M, xlab = "Month", ylab = "Revenue", col= "red", border = "blue", main = "Bar Plot Example")
```

### Stacked Bar Chart
Since more than two variables are represented as a matrix, we create bar chart with
groups of bars and stacks in each bar by using a matrix as input values.

```r
#Example : A monthly revenue of the company respect to regions
#Creating the input vectors.
colors=c("green", "white", "orange")
months=c("Mar", "Apr", "May", "Jun", "Jul")
regions=c("East", "West", "North")
# Create the matrix of the values for stacked barchart
Values=matrix(c(1,9,3,12,9,4,8,7,3,12,5,2,8,10,11), nrow = 3, ncol = 5, byrow = TRUE)
# Create the bar chart with y axis is actually cumulative revenue
barplot(Values, main = "Stacked Bar Plot Exmaple", names.arg = months, xlab = "Month", ylab =
"Revenue", col = colors)
```

### Adding legend to the chart (any)
We use `legend()` function for adding legend to the any chart that we have plotted.  
Syntax is `legend(position, legend vector, cex=size of the legned box, fill=color vector)`

```r
#Adding legend to the above example
#Example : A monthly revenue of the company respect to regions
#Creating the input vectors.
colors=c("green", "white", "orange")
months=c("Mar", "Apr", "May", "Jun", "Jul")
regions=c("East", "West", "North")
# Create the matrix of the values for stacked barchart
Values=matrix(c(1,9,3,12,9,4,8,7,3,12,5,2,8,10,11), nrow = 3, ncol = 5, byrow = TRUE)
# Create the bar chart with y axis is actually cumulative revenue
barplot(Values, main = "Stacked Bar Plot Exmaple", names.arg = months, xlab = "Month", ylab = "Revenue", col = colors)
legend("topleft", regions, cex=1, fill=colors) #legend
```
