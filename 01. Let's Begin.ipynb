### print()
print() is used to print values in R, for example print(“Jai Hind”) will display Jai
hind in the next line. it prints only one object at a time. Trying to print multiple items
gives error message. That is print(“The zero occurs at”, 2 * pi) will give an error message.
The only way to print multiple items is to print them one at a time. There fore above task
can be broken as print(“The zero occurs at”) and print(2 * pi).

### cat()
The cat() function is an alternative to print and concatenate multiple items into
a continuous output.
Example cat(“The zero occurs at”, 2 ∗ pi, “radians.”, “\n”) gives output The zero occurs at
6.28 radians. Note that cat puts a space between each item by default. You must provide a
newline character (\n) to terminate the line.

### ls()
The ls() function displays the names of objects in the workspace.

```r
#Example:
x=−10
z=c("three", "blind", "mice")
ls()
```

Note that ls returns a vector of character strings in which each string is the name of one
variable or function. When your workspace is empty, ls returns an empty vector, which
produces the output:  
**input:** ls()  
**output:** character(0)  

### rm()
The rm() function removes, permanently, one or more objects from the workspace

```r
#Example:
rm(x) #removing x
ls() #checking workspace for x
```

The out put is only 'z', 'x' has been removed from the workspace.

### Rules of naming a variable in R  
1. Names must start with a letter or a dot. If you start a name with a dot, the second character cannot be a digit.
2. Names should contain only letters, numbers, underscore characters ( ) and dots (.). Although you can force R to accept other characters in names, you shouldn’t, because these characters often have a special meaning in R.
3. You cannot use the special keywords as names.

**Note:** R is case sensitive, which means that, for R, lastname and Lastname are two different
objects.

### Comments in R and their uses:
A comment in R is a text followed by #. R ignores everything that appears after
the hash symbol. It can be use the hash symbol at the beginning of a line or somewhere
in the middle. Comments enhance understanding of program. They are usually drafted to
explain what a single line of code does or what it is supposed to produce so that it can help
someone referring to the source code.
